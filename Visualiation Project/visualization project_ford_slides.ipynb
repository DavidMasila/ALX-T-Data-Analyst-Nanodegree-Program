## Investigation Overview

In this investigation i  wanted to know the demographic characteristics of participants in the ford bike share program. 
I also wanted to know what factors influencing participation in the program, be it age, gender, time and and how they affected the program and riders in general. 

## Dataset Overview

This data set includes information about individual rides made in a bike-sharing system covering the greater San Francisco Bay area

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df=pd.read_csv('clean_fordbike.csv')
```

# Number of rides by hour of day. 

Most of the rides in the bike share program are taken at 5pm and 8 am. These are the peak hours when traffic is highest


```python
plt.figure(figsize=[15,8])
sns.countplot(data=df, x="start_hour",color=sns.color_palette()[0])
plt.xlabel("Time of day (24hrs)")
plt.ylabel("Number of rides")
plt.title("Ride traffic by time of day")
plt.show()
```

## Number of Bicycle rides per day of Week

Thursday has the highest number of rides compared to any other day of the week. Overally, weekdays seem to have the most rides while weekends have the least ride numbers

```python
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
plt.figure(figsize=[15,8])
sns.countplot(data=df, x="start_week_day", order=days,color=sns.color_palette()[0])
plt.title("Number of rides per day of week")
plt.xlabel("Week day")
plt.ylabel("Number of rides")
plt.xticks(rotation=30)
plt.show()
```

## Distribution of rider numbers by gender

The number of male riders in the program was by far the largest compared to female and others gender types.

```python
plt.figure(figsize=[15,8])
sns.countplot(data=df, x="member_gender",order=df["member_gender"].value_counts().index, color=sns.color_palette()[0])
plt.title("Number of riders for each gender")
plt.xlabel("Gender")
plt.ylabel("Number of riders")
plt.show()
```

## Rider numbers per age group and day of the week

Thursday is when most of the age groups go for rides. Followed by Tuesday, Friday, Wednesday, Monday and the weekends respectively.

People in their thirties are very active in the bike share program. Their daily riders are by far more compared to the rest of the age groups

```python
ordered=['twenties','thirties','forties','fifties','sixties','oldies']
plt.figure(figsize=[15,8])
sns.countplot(data=df, x='Age_group', hue='start_week_day',order=ordered)
plt.ylabel("Number of riders")
plt.xlabel("Age group")
plt.title("Number of riders by Age_group and day of week")
plt.show()
```

## Duration of rides by Age and user type 

Majority of the customer user type ride for longer durations compared to subcriber user type over all age ranges.

```python
plt.figure(figsize=(15,8))
sns.scatterplot(data=df,x="Age", y="duration", hue="user_type")
plt.title('User_type distribution with regards to age and duration in hours')
plt.xlabel("Age")
plt.show()
```
